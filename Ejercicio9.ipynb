```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

x = np.array([
    14.1, 13.6, 9.3, 8.5, 2.3, 9.8, 3.5, 2.2, 1.7, 1.0, 3.9, 2.4, 4.7, 8.8, 2.5,
    6.4, 8.7, 7.5, 0.7, 8.2, 6.3, 8.4, 4.1, 1.5, 9.6, 6.6, 0.9, 8.5, 5.4, 4.3,
    4.2, 3.3, 0.5, 3.3, 1.0, 9.5, 4.8, 4.0, 8.6, 4.3, 3.0, 0.2, 3.2, 3.3
])
y = np.array([
    5.4, 5.4, 3.5, 5.6, 1.6, 0.3, 5.2, 5.4, 5.7, 5.3, 1.5, 5.5, 2.7, 5.5, 5.6,
    5.5, 5.6, 5.1, 5.6, 5.1, 5.5, 5.6, 1.5, 1.5, 5.3, 5.3, 2.3, 5.3, 5.2, 3.5,
    3.3, 5.3, 5.5, 1.5, 5.3, 5.5, 3.3, 5.5, 1.5, 5.5, 5.5, 5.5, 4.5, 5.3
])

# Ajustar el polinomio de grado 5
p = Polynomial.fit(x, y, deg=5)
coeffs = p.convert().coef

# Mostrar ecuación
print("P(x) = {:.4f} + {:.4f}x + {:.4f}x² + {:.4f}x³ + {:.4f}x⁴ + {:.4f}x⁵".format(*coeffs))

# Graficar
x_plot = np.linspace(min(x), max(x), 200)
y_plot = p(x_plot)

plt.scatter(x, y, color='orange', label="Datos")
plt.plot(x_plot, y_plot, color='red', label="Polinomio grado 5")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación polinómica (grado 5)")
plt.legend()
plt.grid(True)
plt.show()
```
